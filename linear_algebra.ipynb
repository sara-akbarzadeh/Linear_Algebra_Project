# Rotate image

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def rotate_image(image, angle):
    height, width, channels = image.shape
    rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
    return rotated_image

input_image = cv2.imread(r"C:\Users\Sara\Desktop\LA project\LA.jpg")
rotation_angle_45 = 45
rotation_angle_90 = 90

# Rotate the image by 45 degrees
rotated_image_45 = rotate_image(input_image, rotation_angle_45)

# Rotate the image by 90 degrees
rotated_image_90 = rotate_image(input_image, rotation_angle_90)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title(f'Rotated Image ({rotation_angle_45} degrees)')
plt.imshow(cv2.cvtColor(rotated_image_45, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title(f'Rotated Image ({rotation_angle_90} degrees)')
plt.imshow(cv2.cvtColor(rotated_image_90, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

# Image filtering

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_filter(image, kernel):
    filtered_image = cv2.filter2D(image, -1, kernel)
    return filtered_image

input_image = cv2.imread(r"C:\Users\Sara\Desktop\LA project\LA.jpg")
blur_kernel = np.ones((5, 5), np.float32) / 25.0  # 5x5 averaging filter for blurring
sharpen_kernel = np.array([[-1, -1, -1],
                           [-1, 9, -1],
                           [-1, -1, -1]])  # Sharpening filter
edge_detection_kernel = np.array([[-1, -1, -1],
                                  [-1, 8, -1],
                                  [-1, -1, -1]])  # Edge detection filter

blurred_image = apply_filter(input_image, blur_kernel)
sharpened_image = apply_filter(input_image, sharpen_kernel)
edge_detected_image = apply_filter(input_image, edge_detection_kernel)

plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title('Blurred Image')
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 4, 3)
plt.title('Sharpened Image')
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 4, 4)
plt.title('Edge Detected Image')
plt.imshow(cv2.cvtColor(edge_detected_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

# Color manipulation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def convert_to_grayscale(image):
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return cv2.cvtColor(grayscale_image, cv2.COLOR_GRAY2RGB)

def invert_colors(image):
    inverted_image = cv2.bitwise_not(image)
    return inverted_image

def adjust_color_balance(image, alpha=1.0, beta=0.0):
    adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
    return adjusted_image

input_image = cv2.imread(r"C:\Users\Sara\Desktop\LA project\LA.jpg")
grayscale_image = convert_to_grayscale(input_image)
inverted_image = invert_colors(input_image)
adjusted_image = adjust_color_balance(input_image, alpha=1.2, beta=30)

plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 4, 2)
plt.title('Grayscale Image')
plt.imshow(cv2.cvtColor(grayscale_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 4, 3)
plt.title('Inverted Colors')
plt.imshow(cv2.cvtColor(inverted_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 4, 4)
plt.title('Color Balance Adjustment')
plt.imshow(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

# Some other ideas

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

input_image = cv2.imread(r"C:\Users\Sara\Desktop\LA project\LA.jpg")

# Function to blend two images
def blend_images(image1, image2, alpha=0.5):
    blended_image = cv2.addWeighted(image1, alpha, image2, 1 - alpha, 0)
    return blended_image

# Function to perform perspective transformation
def perspective_transform(image, perspective_matrix):
    transformed_image = cv2.warpPerspective(image, perspective_matrix, (image.shape[1], image.shape[0]))
    return transformed_image

# Function for histogram equalization
def histogram_equalization(image):
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    equalized_image = cv2.equalizeHist(grayscale_image)
    equalized_image = cv2.cvtColor(equalized_image, cv2.COLOR_GRAY2RGB)
    return equalized_image

# Function to convert to a different color space
def convert_color_space(image, color_space=cv2.COLOR_BGR2HSV):
    converted_image = cv2.cvtColor(image, color_space)
    return converted_image

# Function for erosion and dilation
def morphological_operations(image):
    kernel = np.ones((5, 5), np.uint8)
    erosion = cv2.erode(image, kernel, iterations=1)
    dilation = cv2.dilate(image, kernel, iterations=1)
    return erosion, dilation

# Function to apply the Sobel operator for edge detection
def sobel_edge_detection(image):
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    sobel_x = cv2.Sobel(grayscale_image, cv2.CV_64F, 1, 0, ksize=5)
    sobel_y = cv2.Sobel(grayscale_image, cv2.CV_64F, 0, 1, ksize=5)
    magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
    return magnitude

# Function for image compression using Run-Length Encoding (RLE)
def run_length_encoding(image):
    flattened_image = image.flatten()
    encoded_image = []
    count = 1

    for i in range(1, len(flattened_image)):
        if flattened_image[i] == flattened_image[i - 1]:
            count += 1
        else:
            encoded_image.extend([flattened_image[i - 1], count])
            count = 1

    encoded_image.extend([flattened_image[-1], count])
    return np.array(encoded_image)

plt.figure(figsize=(15, 15))

# Blending
blended_image = blend_images(input_image, cv2.flip(input_image, 1))
plt.subplot(3, 3, 1)
plt.title('Blended Image')
plt.imshow(cv2.cvtColor(blended_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Perspective Transformation
perspective_matrix = np.float32([[1, 0, 0.2], [0, 1, 0], [0, 0, 1]])
transformed_image = perspective_transform(input_image, perspective_matrix)
plt.subplot(3, 3, 2)
plt.title('Perspective Transformation')
plt.imshow(cv2.cvtColor(transformed_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Histogram Equalization
equalized_image = histogram_equalization(input_image)
plt.subplot(3, 3, 3)
plt.title('Histogram Equalization')
plt.imshow(cv2.cvtColor(equalized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Color Space Conversion
hsv_image = convert_color_space(input_image, cv2.COLOR_BGR2HSV)
plt.subplot(3, 3, 4)
plt.title('HSV Color Space')
plt.imshow(cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB))
plt.axis('off')

# Morphological Operations
erosion, dilation = morphological_operations(hsv_image[:, :, 2])
plt.subplot(3, 3, 5)
plt.title('Erosion')
plt.imshow(erosion, cmap='gray')
plt.axis('off')
plt.subplot(3, 3, 6)
plt.title('Dilation')
plt.imshow(dilation, cmap='gray')
plt.axis('off')

# Sobel Edge Detection
edge_image = sobel_edge_detection(input_image)
plt.subplot(3, 3, 7)
plt.title('Sobel Edge Detection')
plt.imshow(edge_image, cmap='gray')
plt.axis('off')

# Run-Length Encoding Compression
compressed_image = run_length_encoding(edge_image)
plt.subplot(3, 3, 8)
plt.title('Run-Length Encoding Compression')
plt.plot(compressed_image)
plt.axis('off')

plt.show()
```
